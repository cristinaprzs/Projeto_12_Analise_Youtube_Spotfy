### 5. Quais as 5 músicas com musicas com mais views no youtube?

```python
### quero mostrar as 5 musicas com mais views no youtube
import matplotlib.pyplot as plt

top5views = df.sort_values(by="Views", ascending=False).head(5)
plt.figure(figsize=(8,5))
plt.bar(top5views["Track"], top5views["Views"])
plt.show()
```

### 6. Quais os 5 artistas com mais streams no spotify?

```python
### Quais os 5 artistas com mais streams no spotify?
top5_streamsm = df.groupby("Artist")
["Stream"].sum().sort_values(ascending=False).head(5)

print(top5_streams)
```
